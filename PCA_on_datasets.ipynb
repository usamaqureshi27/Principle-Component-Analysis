# Principle Component Analysis (PCA)

## PCA on IRIS dataset

```python
# Import the libraries 
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
iris = load_iris()
```

```python
# Reduce the data into three dimensions
pca = PCA(n_components=3)
```

```python
# Fit and trasnform the data
iris_pca = pca.fit_transform(iris.data)
```

```python
# Plot the first two principle components
plt.scatter(iris_pca[:,0] , iris_pca[:,1] , c=iris.target)
plt.xlabel('PC1')
plt.ylabel('PC1')
plt.show
```

```python
# Plot the second and third principle components
plt.scatter(iris_pca[:,1] , iris_pca[:,2] , c=iris.target)
plt.xlabel('PC2')
plt.ylabel('PC3')
plt.show
```

## PCA on TIPs dataset

```python
# Import the libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
```

```python
# Load the data
tips = sns.load_dataset('tips')
```

```python
# Preprocessing
# encode categorical data

le = LabelEncoder()

cat_feature = tips.select_dtypes(include='category').columns

for features in cat_feature:
    tips[features] = le.fit_transform(tips[features])
    
# standardize the data

scaler = StandardScaler()
tips_scaled = scaler.fit_transform(tips)
```

```python
# Apply PCA
pca = PCA() # Reduce to k=10 dimensions with PCA
```

```python
# Fit and trasnform the data
X_pca = pca.fit_transform(tips_scaled)
```

```python
# Create a Dataframe to show the loadings
loadings = pd.DataFrame(pca.components_.T, columns = [f'PC{i+1}' for i in range(pca.n_components_)] , index = tips.columns) 
print(loadings)
```
